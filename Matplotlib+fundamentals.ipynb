```python
from matplotlib.artist import Artist

def rec_gc(art, depth=0):
    if isinstance(art, Artist):
        # increase the depth for pretty printing
        print("  " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

# Call this function on the legend artist to see what the legend is made up of
rec_gc(plt.legend())
```

```python
l_data = np.array([1,2,3,4,5,6,7,8])
e_data = l_data**2

plt.figure()
plt.plot(l_data, '-o', e_data, '-o')
```

```python
plt.gca().fill_between(range(len(l_data)), 
                       l_data, e_data, 
                       facecolor='green', 
                       alpha=0.25)
```

```python
plt.figure()

o_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

plt.plot(o_dates, l_data, '-o',  o_dates, e_data, '-o')
```

```python
import pandas as pd

plt.figure()

o_dates = np.arange('2017-01-01', '2017-01-09', dtype = 'datetime64[D]')
o_dates = list(map(pd.to_datetime, o_dates))
plt.plot(o_dates, l_data, '-o', o_dates, e_data, '-o')
```
