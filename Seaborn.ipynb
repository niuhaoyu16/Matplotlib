```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib notebook
```

```python
np.random.seed(1234)

a = pd.Series(np.random.normal(0,10,1000), name = 'a')
b = pd.Series(2*a + np.random.normal(60,15,1000), name = 'b')
```

```python
plt.figure()
plt.hist(a, alpha=0.7, bins=np.arange(-50,150,5), label = 'a');
plt.hist(b, alpha=0.7, bins=np.arange(-50,150,5), label = 'b');
plt.legend();
```

```python
plt.figure()
plt.hist([a, b], histtype='barstacked', normed  = True);
c = np.concatenate((a,b))
sb.kdeplot(c);
```

```python
plt.figure()
sb.distplot(c, hist_kws = {'color': 'Teal'}, kde_kws= {'color':'Navy'});
```

```python
sb.jointplot(a,b,alpha=0.4);
```

```python
grid = sb.jointplot(a, b, alpha=0.4);
grid.ax_joint.set_aspect('equal')
```

```python
sb.jointplot(a,b,kind='hex');
```

```python
sb.set_style('white')
sb.jointplot(a,b,kind='kde',space=0);
```
